```python
import pandas as pd
```

```python
data=pd.read_csv("test_data_with_predictions.csv")
```

```python
data.columns
```

```python
data.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plotting the SalesPrediction column
plt.figure(figsize=(12, 6))
sns.histplot(data=data, x='SalesPrediction', bins=30, kde=True, color='skyblue')
plt.title('Distribution of Sales Predictions')
plt.xlabel('Sales Predictions')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Example: Bar plot for 'StoreType' counts
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='StoreType', palette='Set2')
plt.title('Store Type Distribution')
plt.xlabel('Store Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Example: Line plot for 'SalesPrediction' over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='Date', y='SalesPrediction', color='purple')
plt.title('Sales Prediction Over Time')
plt.xlabel('Date')
plt.ylabel('Sales Prediction')
plt.xticks(rotation=45)
plt.show()
```

```python
# Example: Correlation heatmap for numerical columns
correlation_matrix = data.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Example: Pairplot for numerical columns
sns.pairplot(data[['SalesPrediction', 'CompetitionDistance', 'Promo', 'SchoolHoliday']], diag_kind='kde', palette='Set2')
plt.title('Pairplot of Numerical Features')
plt.show()
```

```python
# Example: Violin plot for 'StoreType' vs. 'SalesPrediction'
plt.figure(figsize=(12, 6))
sns.violinplot(data=data, x='StoreType', y='SalesPrediction', palette='Set3')
plt.title('Sales Prediction Distribution by Store Type')
plt.xlabel('Store Type')
plt.ylabel('Sales Prediction')
plt.xticks(rotation=45)
plt.show()
```

```python
# Example: Bar plot for 'Season' vs. 'SalesPrediction'
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='Season', y='SalesPrediction', palette='pastel')
plt.title('Sales Prediction by Season')
plt.xlabel('Season')
plt.ylabel('Sales Prediction')
plt.show()
```

```python
# Example: Box plot for 'SalesPrediction' by 'DayOfWeek'
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='DayOfWeek', y='SalesPrediction', palette='Set2')
plt.title('Sales Distribution by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Sales Prediction')
plt.show()

```

```python
# Example: Distribution plot for 'CompetitionDistance'
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='CompetitionDistance', bins=30, kde=True, color='lightgreen')
plt.title('Distribution of Competition Distance')
plt.xlabel('Competition Distance')
plt.ylabel('Frequency')
plt.show()
```

```python
# Example: Pie chart for 'StoreType' distribution
store_type_counts = data['StoreType'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(store_type_counts, labels=store_type_counts.index, autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral', 'lightsalmon'])
plt.title('Store Type Distribution')
plt.show()
```

```python
# Example: Scatter plot for 'CompetitionOpenSinceYear' vs. 'SalesPrediction'
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='CompetitionOpenSinceYear', y='SalesPrediction', color='skyblue')
plt.title('Sales Prediction vs. Competition Open Since Year')
plt.xlabel('Competition Open Since Year')
plt.ylabel('Sales Prediction')
plt.show()
```

```python
# Example: Histogram of 'SalesPrediction' on school holidays vs. non-school holidays
plt.figure(figsize=(10, 6))
sns.histplot(data=data[data['SchoolHoliday'] == 1]['SalesPrediction'], bins=30, kde=True, label='School Holiday', color='darkblue')
sns.histplot(data=data[data['SchoolHoliday'] == 0]['SalesPrediction'], bins=30, kde=True, label='Non-School Holiday', color='lightgreen')
plt.title('Sales Distribution on School Holidays vs. Non-School Holidays')
plt.xlabel('Sales Prediction')
plt.legend()
plt.show()
```

```python
# Example: Bar plot for 'Promo2' vs. 'SalesPrediction'
plt.figure(figsize=(8, 6))
sns.barplot(data=data, x='Promo2', y='SalesPrediction', palette='Set3')
plt.title('Sales Prediction for Promo2 vs. Non-Promo2 Stores')
plt.xlabel('Promo2')
plt.ylabel('Sales Prediction')
plt.show()
```
